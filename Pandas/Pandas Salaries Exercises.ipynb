Import pandas as pd

````python
import pandas as pd
````

Read Salaries.csv as a dataframe called 'sal'

````python
sal=pd.DataFrame(pd.read_csv('Salaries.csv'))
````

Check the head of the DataFrame

````python
sal.head()
````

Use the ```.info()``` method to find out how many entries there are.

````python
sal.info()
````

What is the average BasePay?

````python
sal['BasePay'].mean()
````

What is the highest amount of OvertimePay in the dataset?

````python
sal['OvertimePay'].max()
````

What is the job title of JOSEPH DRISCOLL?

*Note: use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll)*

````python
sal.loc[sal['EmployeeName']=='JOSEPH DRISCOLL','JobTitle']
````

How much does JOSEPH DRISCOLL make (including benifts)?

````python
sal.loc[sal['EmployeeName']=='JOSEPH DRISCOLL','TotalPayBenefits']
````

What is the name of the highest paid person(including benefits)?

do you notice something strange about how much he or she is paid?

````python
sal.loc[sal['TotalPayBenefits']== sal['TotalPayBenefits'].min()]
````

What was the average (mean) BasePay of all employees per year (2011-2014)?

````python
sal.groupby('Year')['BasePay'].mean()
````

How many unique job titles are there?

````python
sal['JobTitle'].nunique()
````

What are the top 5 most common jobs?

````python
sal.groupby('JobTitle')['JobTitle'].count().sort_values(ascending=False).head()
````

How many job titles were represented by only one person in 2013?

*(e.g. JobTitles with only one occurance in 2013)*

````python
sal.loc[(sal['Year']==2013)].groupby('JobTitle')['JobTitle'].filter(lambda x: x.count() ==1).count()
````

How many people have the word "Chief" in their job title?

````python
sal.loc[(sal['JobTitle'].str.contains('Chief', regex=False,case=True)),'JobTitle'].count()
````
