```python
import seaborn as sns
%matplotlib inline
```

```python
#seaborn comes with some sample pre-built data sets. so using one of them-tips
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distplot--plot for univariant set of observations, means only a single variable. the curve is known as kde. to make it appear
#mention kde=True(by default). bins=no od blocks. try using bin=10. with dist plot we get histogram, where x axis is total_bill
#and y axis has the count/number of bills/frquency. which means most of the total_bill lies between 10 and 20(10 and 20 on x axis)
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
#jointplot--allows you to match up distplot of bivariant data, which means you can join 2 distplots
sns.jointplot(x='total_bill',y='tip',data=tips)
#reading the jointplot below--the higher the bill the higher the tip as suggested by the scatter plot
```

```python
#jointplot gives argument 'kind'.kind=hex--hexagonal distribution.hexagon will be dark where count is more
#kind=reg--regression line in the scatter plot
#kind=kde--2-D kde gets generated
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#pairwise realtionship across entire dataframe.means jointplot for every single numeric columns in this dataframe
#whereever histograms are there, its total_bill vs total_bill, tip vs tip and size vs size
#has parameters like hue=column name of a categorical column
#palette='coolwarm'--different colors
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#rugplot--pass a single column. It just draws a dash along this univariant
sns.rugplot(tips['total_bill'])
```

```python
#compare rugplot with distplot
#histogram has bins, it counts the number of occurences/number of dashes in that bin and shows it along the y axis(refer the top of rectangles in histogram)
sns.distplot(tips['total_bill'],kde=False)
```

# Categorical Plots

```python
tips.head()
```

```python
#in categorical plots we will look after the distribution between a categorical column(like sex in this case) vs a numeric or any other categorical column
#barplot--aggregate categorical data based off some function, by default that function is mean(). so this shows mean of total bill for each categorical value
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
#change the estimator object to change the grouping in barplot
#estimator is a function that is an aggregate function(check docstring). you can use custom defined aggregate functions also.here we mention standard deviation(np.std)
#tells the standard deveiation of total_bill per category
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#countplot--similar to barplot. here the estimator counts the number of occurences and thus we just set the x value. estimator becomes the y axis
sns.countplot(x='sex',data=tips)
```

```python
#box plot and violin plots--shows the distribution of quantative data facilating comparison between variables
#distribution for total bill per day. box plot shows the quartiles and outliers
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#split box plots by another categorical column using hue parameter
#as per the plot,nonsmokers have a larger bill on Fridays as comapred to smokers
#check 'orient' parameter for boxplot
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#violin plots--not much used as it is difficult to present and make someone understand the meaning of it. A data scientist will beable to understand not any body else
#split=True--splits the violin plot into Male and Female depending upon the hue column
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#strip plot--draws a scatter plot where one variable is categorical
#add jitter=True parameter to add some random noise so as to make this plot look better
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True)
```

```python
#swarmplot--points are adjusted so that they dont overlap. scatter plot strip plot and violin plot combined
sns.swarmplot(x='day',y='total_bill',data=tips)
```
