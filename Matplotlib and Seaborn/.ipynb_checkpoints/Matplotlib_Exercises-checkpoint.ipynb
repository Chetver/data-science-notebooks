# Matplotlib Exercises

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
%matplotlib inline
````

## Data:

````python
x = np.arange(0,100)
y = x*2
z = x**2
````

### Exercise 1
Follow along with these steps:
- Create a figure object called fig using ```plt.figure()```
- Use ```add_axes``` to add an axis to the figure canvas at [0,0,1,1]. Call this new axis 'ax'
- Plot (x,y) on that axis and set the labels and titles (title=titles, x axis = x, y axis = y)

````python
fig1=plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
````

## Exercise 2
Create a figure object and put two axes on it, ```ax1``` and ```ax2```, located at ```[0,0,1,1]``` and ```[0.2,0.5,.2,.2]``` respectively.

````python
fig2=plt.figure()
ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([0.2,0.5,.2,.2])
````

Now plot (x,y) on both axes and call your figure object to show it.

````python
ax1.plot(x,y)
ax2.plot(x,y)
fig2
````

## Exercise 3
Create the plot below by adding two axes to a figure object at ```[0,0,1,1]``` and ```[0.2,0.5,.4,.4]```.

````python
fig3=plt.figure()
ax1 = fig3.add_axes([0,0,1,1])
ax2 = fig3.add_axes([0.2,0.5,.4,.4])
````

Now use x,y and z arrays to recreate the plot below. Notice the x limits and y limits on the inserted plot.

````python
ax1.plot(z,'b')
ax1.set_ylabel('z')
ax2.plot(x,y,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_ylim([30, 50])
ax2.set_xlim([20, 22])
fig3
````

## Exercise 4
Use ```plt.subplots(nrows=1, ncols=2)``` to create the plot below.

````python
fig4,axes4 = plt.subplots(nrows=1, ncols=2)
````

Now plot (x,y) and (x,z) on the axes.

````python
axes4[0].plot(x,y,'r')
axes4[1].plot(x,z,'r')
fig4
````
