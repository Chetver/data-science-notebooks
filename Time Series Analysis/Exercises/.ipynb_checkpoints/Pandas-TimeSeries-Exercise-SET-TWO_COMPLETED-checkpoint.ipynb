**Plot out the yearly rolling mean on top of the original data. Recall that this is monthly data and there are 12 months in a year!**

```python
# CODE HERE
#df.drop(['Yearly Mean'],inplace=True,axis=1)
df_copy = df.copy()
df_copy['Yearly Mean'] = df_copy['UMTMVS'].rolling(window=12).mean()
df_copy.plot(figsize=(12,5)).autoscale(axis='x',tight=True);
```

**BONUS QUESTION (HARD).**

**Some month in 2008 the value peaked for that year. How many months did it take to surpass that 2008 peak? (Since it crashed immediately after this peak) There are many ways to get this answer. NOTE: I get 70 months as my answer, you may get 69 or 68, depending on whether or not you count the start and end months.

```python
# Starting year to search within:
search_year = '2008'
############################

peak_index = df.loc[search_year].idxmax() #Find the index for the maximum value for the year
peak_value = df.loc[peak_index,'UMTMVS'].values[0].astype(float) #Store the maximum value for the initial peak

# For each index in the DataFrame that is after the initial peak
for idx in df.truncate(before=peak_index[0],axis=0,copy=True).index:
    
    # check if the value is higher than the initial peak value
    if(df.loc[idx,'UMTMVS']>peak_value):
        next_peak_index = idx
        next_peak_value = df.loc[idx,'UMTMVS']
        # Calculate the number of months between the initial peak and the next peak
        next_peak_months = ((next_peak_index - peak_index) / np.timedelta64(1, 'M')).values.astype(int)
        break

# Explain the results in a meaningful way
print("In "+str(search_year)+", the value peaked in "+peak_index[0].month_name()+" at $"+str(peak_value)+".\nIt took "+str(next_peak_months[0])+" months, until "+str(next_peak_index.month_name())+" of "+str(next_peak_index.year)+", for the value to surpass the previous peak and reach $"+str(next_peak_value)+".")
```
