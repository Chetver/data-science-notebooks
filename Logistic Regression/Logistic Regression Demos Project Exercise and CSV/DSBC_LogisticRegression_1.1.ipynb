# Summary Explanation:

### Maximum Likelihood Estimation(MLE): Method used
it is based on the likelihood function which is- "a function which estimates how likely it is that the model at hand describes the real underlying relationship of the variables". In simpler terms, "the bigger the likelihood function, the higher the probability that our model is correct".
MLE tries to maximize the likelihood function and hence the name.
    What happens at the backend-- the computer goes through different values until it finds a model, for which the likelihood is the highest. When it can no longer improve it, it stops the optimization
    
### Log-likelihood:
It is much more convenient to take the log-likelihood when performing MLE. So the explanation remains the same as above, with just an addition of the term 'log':
MLE tries to maximize the 'log' likelihood function.
    What happens at the backend-- the computer goes through different values until it finds a model, for which the 'log' likelihood is the highest. When it can no longer improve it, it stops the optimization
    The value of log-likelihood is almost but not always negative and the bigger it is the better
    
### Log-likelihood Null(LL-Null):
The log likelihood of a model which has no independent variables. The sole independent variable is just an array of 1s (y=b0*x0-- here x0 is 1 and thus we get y=b0+b1x1+b2x2 and so on)
This array is the constant we are adding with the add_constant(). You can try doing that. If you have an array of ones for x1 and try to use the logit function and build a logistic regression model, then this model's 'log-likelihood' will be equal to the LL-Null value of the results_log model

### LLR p-value:
You may want to compare the log-likelihood of your model with LL-Null, to see if your model has any explanatory power. Recall that for linear regression we used the F-statsitic to determine the significance of our model. In Logistic regression , this is achieved by LLR(Log likelihood ratio test). It is based on the log-likelihood of the model and the LL-Null. It measures if our model is statistically different from LL-NUll,i.e a useless model. You can see that the p-value is very low =0.000 and thus our model is significant

### Pseudo R-squared:
Also known as McFadden's R-squared. A good value is somewhere between 0.2 to 0.4. Mostly used for comparing the variations of the same model. Different models will have completely different Pseudo R-squares

## Coefficient table
As per the regression summary our model is as follows:
log(pi/1-pi)=-69.91+0.042*SAT
Lets go through the process of interpreting the result. Lets take 2 SAT scores, SAT1 and SAT2. So now the 2 equations are:
log(odds2)=-69.91+0.042*SAT2
log(odds1)=-69.91+0.042*SAT1

Going ahead with the difference of the 2 equations:
log(odds2)-log(odds1)=0.042(SAT2-SAT1), which can also be written as:

log(odds2/odds1)=0.042(SAT2-SAT1)

Now lets assume that the difference between the 2 SAT scores is 1. Substituing the values, we get:
log(odds2/odds1)=0.042. Now lets take exponential of the 2 sides(recall that exponential of log(x)=x):

Taking exponential on both sides we get:
odds2/odds1=1.042

which means that odds2 is 1.042*odds1 or odds2=104.2%(odds1) or odds2 is 4.2% higher than odds1.
So when the SAT score increases by 1 unit, the odds of admittance increase by 4.2%
General rule-- for a unit change in a variable,the change in the odds equals the exponential of the coefficient(difference of odds=e^b, where b is the coefficient of x)
That exactly is what provides us with a way to interpret the coefficients of a logistic regression.

Similarly the difference between the SAT scores can be 10,12,20,100

log(odds2/odds1)=2.08*(Gender2-Gender1)--Gender has only 2 possible values 1 and 0. So let Gender2=1 and Gender1=0:

log(odds2/odds1)=2.08

```python
def confusion_matrix(data,actual_values,model):
        
        # Confusion matrix 
        
        # Parameters
        # ----------
        # data: data frame or array
            # data is a data frame formatted in the same way as your input data (without the actual values)
            # e.g. const, var1, var2, etc. Order is very important!
        # actual_values: data frame or array
            # These are the actual values from the test_data
            # In the case of a logistic regression, it should be a single column with 0s and 1s
            
        # model: a LogitResults object
            # this is the variable where you have the fitted model 
            # e.g. results_log in this course
        # ----------
        
        #Predict the values using the Logit model
        pred_values = model.predict(data)
        # Specify the bins 
        bins=np.array([0,0.5,1])
        # Create a histogram, where if values are between 0 and 0.5 it will be considered 0
        # if they are between 0.5 and 1, they will be considered 1
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # Calculate the accuracy
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # Return the confusion matrix and the accuracy
        return cm, accuracy
```

```python
# Create a confusion matrix with the test data
cm = confusion_matrix(test_data,test_actual,results_log)
cm
#Left part is the confusion matrix and right part is the accuracy
#Almost always, the training accuracy is higher than the testing accuracy
```

# Market Segmentation with Cluster Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('3.12. Example.csv')
data
#Each observation is a client showcasing the customer satisfaction(rating the shopping exeperience on a scale of 1-10, or how many stars, so its a discrete numeric variable) and loyalty(usually depends upon various factors like churn rate, retention rate. Here it has been already calculated and we have to keep into account that the value ranges from -2.5 to 2.5)
```

```python
#Plotting the data
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
#you can clearly observe 2 clusters as per the plot shown below
```

### the above graph can be divided into 4 sections as follows:
so talking in terms of quadrants in clockwise direction, starting from the top left quadrant:
top-left quadrant(1st quad)-- low satisfaction, high loyalty
top-right quadrant(2nd quad)-- high statisfaction, high loyalty
bottom-right(3rd)--high satisfaction, low loyalty
bottom-left(4th)-- low satisfaction low loyalty
As per this, the2 cluster solution will not make much sense

```python
x=data.copy()
```

```python
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
#Cluster is a duplicate of x
clusters=x.copy()
#column cluster_pred of clusters will contain the predictions
clusters['cluster_pred']=kmeans.fit_predict(x)
```

```python
# Plot the data using the longitude and the latitude
# c (color) is an argument which could be coded with a variable 
# The variable in this case has values 0,1, indicating to plt.scatter, that there are two colors (0,1)
# All points in cluster 0 will be the same colour, all points in cluster 1 - another one, etc.
# cmap is the color map. Rainbow is a nice one, but you can check others here: https://matplotlib.org/users/colormaps.html
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
#asper the above plot, you can observe that on the x-axis, there is cutoff line at 6. Everything to its right is one cluster and to the left is another cluster
#this seems ok, however the problem is that we havent standardized the variable.Observe the scale of Satisfaction and that of Loyalty.
#So kmeans more or less disregarded loyalty as a feature
#whenever we cluster on basis of a single feature we will always obtain these type of results, cutoff by a single line kind
#So we have to give both loyalty and satisfaction equal weight(standardizing satisfaction)
```

## Standradize the variable

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(x)
#preprocessing.scale(x)--scales(standardizes with mean 0 and std of 1 by default) each variable(column) separately with respect to itself
x_scaled
```

```python
 #since we dont know the number of clusters needed, the elbow method will prove useful
wcss=[]

for i in range(1,10):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
```

```python
wcss
```

```python
# Plot the number of clusters vs WCSS
plt.plot(range(1,10),wcss)
# Name your axes
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
#4 clusters seems a good idea!
```

## Explore clustering solutions and select the number of clusters

```python
# Fiddle with K (the number of clusters)
kmeans_new = KMeans(2)
# Fit the data
kmeans_new.fit(x_scaled)
# Create a new data frame with the predicted clusters
clusters_new = x.copy()
#cluster_new has a new column 'cluster_pred' which will contain the predictions of the scaled x
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
clusters_new
#The data frame contains the original values but the predicted clusters are based on the solution using
#the standardized data.We will plot the data without standardizing it,but the solution itself is the standardized one.
```

```python
# Plot
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
#we call the red cluster- the Alienated cluster(less satisfied less loyal)
#other one is everything else cluster
```
