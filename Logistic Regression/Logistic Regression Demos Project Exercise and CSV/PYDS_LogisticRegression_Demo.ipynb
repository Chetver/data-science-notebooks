```python
#fill in mean age of all passengers data--imputation
#first check the average age of passenger per class using boxplot
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass',y='Age',data=train)
#observe that passengers whose age was more belonged to 1st class as compared to the younger ones who belonged to the 3rd class
```

```python
#you can also go ahead and create a model which will predict the age on basis of other features
#however here lets just stick to a simple imputation(filling in). creating a function for this
#with the function we are finding out the average age of people per class. SO rather than just using the mean of all the age values we are using the average age Pclass wise
#check the median of the 1st box plot. passenger class 1,average age 37
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#no null values to be seen now
```

```python
#Cabin is having a lot of null values. so better lets drop it
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
#so data cleansing has been done wherein we have taken care of the missing values.
#rather than just ignoring the records with nulls for specific columns we used the impute function
#execute the undermentioned code again to check the heatmap for any other nulls
#check the docstring for yticklabels and make it to true to understand what is happening
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#dealing with categorical features by converting them into dummy variables using pandas. Else a ML algorithm will not be able to take those features as input
#Sex column has categorical features like male or female. the string of characters cannot be processed properly by a ML algo
#so we have to create a new column wherein the male and female strings can be represented as 0/1 and thus can be understood by ML algo
#we will do the same thing for Embarked column which has S,C, Q as values. this is known as converting to dummy variables
#get_dummies convert categorical variables into dummies--pass the train data and the column we want to convert, in this case Sex column
pd.get_dummies(train['Sex'])
#observe the output.we now have a dataframe with a column for every single category and a 0 or 1 value. 
#both thecolumns male and female have 0 and 1. Examine the first record of train.head(). its a male thus, here the value for that being 1 and similarly coresponding female value is 0
#also one column is perfect predictor for the other column. Meaning,the ML algo can predict like if female has value 0 it means male will be 1
#this is the issue of multi-collinearity as a bunch of columns become perfect predictors for the other columns
```

```python
#to get over the issue of multi-collinearity use drop_first
pd.get_dummies(train['Sex'],drop_first=True)
```

```python
#we want to feed the above data to our ML algo.So we assign it to a df named sex as shown below:
sex=pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex.head()
```

```python
#doing similar for embark
embark=pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
embark.head()
#the C column has been dropped. we just have indicators for Q and S and those are not perfect predictors of each other. For Q=0, S is 1 also and at some places its 0 also
```

```python
#adding the new columns to the train data
train=pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

```python
#now we dont need the Sex and Embarked column. Also we can do some feature engineering on Name and Ticket column like extracting the last name, or the first letter. However, once again, these text columns are not of much use for the ML algo
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train.head()
```

```python
#similarly passenger id is also not of much use because there is no point in extracting aggregated info on passenger id column
#Also we cannot predict like if passengerId is 1,400,801 then the person survived else the person didnt survive
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train.head()
```

```python
#we have titanic_train.csv and titanic_test.csv.We can either clean the test.csv file and use that to evaluate the model
#Here I am just splitting the titanic_train.csv into test and train
#X--the other set of features apart from Survived column, as Survived is the one which we are trying to predict
#y--survived
X=train.drop('Survived',axis=1)
y=train['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#training and prediction by creating a model
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
#call some predictions using the X_test
predictions=logmodel.predict(X_test)
```

```python
#for performing classification tasks
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
#to display the confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
