## Logistical Regression

- Yes/No
- 1/0
- True/False

Step 1: Map catergorical values to '1' and '0'.

No values other than 1 or 0, such that it is nessecary to define 0.5 so that all values above 0.5 are considered to be 1 and all values below 0.5 are considered to be 0.


Linear Regression Model:
$$y=b_0 + b_1 x_1 + b_2 x_2 + b_3 x_3 + \dots b_n x_n$$

Logistical Regression Model:
$$p(x)=\frac{e^{(b_0 + b_1 x_1 + b_2 x_2 + b_3 x_3 + \dots b_n x_n)}}{1+e^{(b_0 + b_1 x_1 + b_2 x_2 + b_3 x_3 + \dots b_n x_n)}} $$

Logic Function:

$$\frac{P(x)}{1-P(x)}=e^{(b_0 + b_1 x_1 + b_2 x_2 + b_3 x_3 + \dots b_n x_n)}$$

$$log\bigg(\frac{P(x)}{1-P(x)}\bigg)= b_0 + b_1 x_1 + b_2 x_2 + b_3 x_3 + \dots b_n x_n$$

To determine the 'odds' of an event, use the logic function:
- 1/6 Chance of event occuring
- 5/6 Chance of event not occuring
$$log\bigg(\frac{1/6}{5/6}\bigg)=1:5$$

Using a confusion matrix:

|Name|Formula|Description|
|:--|:-:|:-------|
**Recall**|$\frac{\text{TP}}{\text{TP+FN}}$|Out of all positive classes, how much we predicted correctly
**Precision**|$\frac{\text{TP}}{\text{TP+FP}}$|Out of all the classes we predicted positive (correct), how many are actually postive
**Accuracy**|$\frac{\text{TP+TN}}{\text{Total}}$|Out of all the predictions, how many were correct (where Total is the complete sample size)
**Misclasification/Error Rate**|$\frac{\text{FP+FN}}{\text{Total}}$|How often the model is wrong

### F-Measure
Used when comparing two models, Typically in the scenario where:

Model 1: $\downarrow$ Recall $\uparrow$ Precision

Model 2: $\uparrow$ Recall $\downarrow$ Precision

$$\text{F-Measure}=\frac{2\times\text{Recall}\times\text{Precision}}{\text{Recall}+\text{Precision}}$$

## Clustering
*Unsupervised Machine Learning*

Multivariate statistical technique that groups observations on basis of some of the features/variables they are described by.

Example:

Countries with the features of Lat, Long, and Language.

The countries could be grouped by Lat,Long or they could be grouped by Language

### K-Means

K-Means$(x)$, where $x$ is the number of clusters.

Steps:
1. - Define the number of clusters
2. - Identify the 'seed' or Centroid
3. - Based on proximity/distance from the indentified points, group the values.

Maximum number if iterations = 35

### Final goal of clustering

To maximise the similarity of classification within a cluster and to maximise the dissimilarity between clusters.

### Clustering Mathematical Concepts

1. Euclidean Distance
    - The distance between two or more points.
    - 2D: $\sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}$
    - 3D: $\sqrt{(x_2-x_1)^2 + (y_2-y_1)^2 + (z_2-z_1)^2}$
    - https://en.wikipedia.org/wiki/Euclidean_distance
2. Centroid
    - The exact center point between $n$ number of points
    - https://en.wikipedia.org/wiki/Centroid
